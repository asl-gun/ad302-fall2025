**The following code is used for a class, in which whether the students pass or fail depends solely on their performance in the midterm and in tne final. The midterm effects the end score by %40 and the final with %60. In order to pass the class, the students must have a score of at least 50.**

```r
set.seed(123)
```

***(1) Creating a dataframe that shows students in a class and their midterm and final notes***

```r
class_notes<-data.frame(
    Students=c("Aslı","Ata","Cemre","Emirhan","Yekta","Tuna","Emir","Ece"),
    midterm=sample(0:100,8,replace=T),
    final=sample(0:100,8, replace=T)
    )
class_notes
```

***(2) Creating a function "grades" that calculate the students' overall grade with respect to the evaluation criteria of the midterm and the final exams.***

```r
grades<-function(data){
  weighted_grades<-0.4*data$midterm + 0.6*data$final
  names(weighted_grades)<-data$Students
  return(weighted_grades)
}

end_notes<-grades(class_notes)
end_notes
```

***(3) Creating another function "pass_or_fail" to determine if the students passed or failed the class. The function checks every students' overall grade with a loop and assigns the result as "pass" or "fail" and returns a list of student names and their pass/fail situation.***

```r
pass_or_fail<-function(score){
  results <- c()
  for (student in names(score)){
    x<-score[[student]]
    if (x>=50){
      results[[student]]<-"pass"
    }else{
      results[[student]]<-"fail"
    }
  } 
  return(results)
}
paf<-(pass_or_fail(end_notes))
paf
```
